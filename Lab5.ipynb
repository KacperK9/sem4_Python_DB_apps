Zadanie 22:
    Opracuj wykres relacyjny z wykorzystaniem biblioteki seaborn, który przedstawi zależność międzykolumną 
    SurvivediAge. 
    Wykres powinien posiadać:tytuł,opis osi X,opis osiY,legendę.
        
    Ponadto, proszę samodzielnie wybrać wymiary wykresu oraz proszę wyeksportować wykres do pliku w formacie PNG (nazwa dowolna).

```python
titan=pd.read_csv('titanic.csv')

plot = sb.catplot(data=titan, x='Survived', y='Age')
plot.set_xlabels('wiek')
plot.set_ylabels('ocaleni')
plt.title('Wykres zależności', fontsize=16)
plt.savefig('titanic.png', dpi=250)
```

Zadanie 23:
Opracuj funkcję, której zadaniem będzie wyświetlenie wykresu relacyjnego z wykorzystaniem biblioteki seaborn, który przedstawi zależność między kolumną Survivedi Age uwzględniając wartości w kolumnie Pclass.

```python
titan=pd.read_csv('titanic.csv')

plot = sb.catplot(data=titan, hue='Pclass', x='Survived', y='Age')
plot.set_xlabels('wiek')
plot.set_ylabels('ocaleni')
plt.title('Wykres zależności z uwzględnieniem Pclass', fontsize=16)
plt.savefig('titanic_pclass.png', dpi=250)
```

Opracuj wykres (dwie funkcje na jednym wykresie) funkcji trygonometrycznych y=sin(x) i y=cos(x) w przedziale<-4;4> z wykorzystaniem biblioteki seaborn.Ponadto, proszę samodzielnie wybrać wymiary wykresu oraz proszę wyeksportować wykres do pliku w formacie PNG (nazwa dowolna).

```python
import math
pi=math.pi
sinwave = np.sin(np.arange(-4*pi, 4*pi, 0.1))
coswave = np.cos(np.arange(-4*pi, 4*pi, 0.1))

plt.plot(sinwave)
plt.plot(coswave)
plt.savefig('tryg.png', dpi=250)
```
